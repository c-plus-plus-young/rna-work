```python
# THIS IS FOR EXPERIMENTS WITH MULTIPLE FILES, WHERE EACH FILE IS A SINGLE SAMPLE

# Initialize an empty list to store DataFrames
dfs = []

gene_column = 'Gene'
count_column = 'Count'
# Loop through all TSV files in the folder
for file_name in os.listdir(folder):
    if file_name.endswith('.tab'):
        file_path = os.path.join(folder, file_name)
        
        # Read the TSV file
        df = pd.read_csv(file_path, sep='\t', skiprows=1).iloc[:, [0, 1]]
        df.columns = ['Gene', 'Count']

        # df = pd.read_csv(file_path, sep='\t', skiprows=0)[[gene_column, count_column]]

        # df = pd.read_csv(file_path, sep='\t', skiprows=0).iloc[:, [0, -1]]
        
        # Set 'Id' as the index and rename 'Count' to the file name (without extension)
        df.set_index(gene_column, inplace=True)
        filename = '_'.join(file_name.replace('.txt', '').split('_')[1:3]).split('.')[0].replace('-', '_')
        df.rename(columns={count_column: filename}, inplace=True)
        
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames along the columns, filling missing IDs with 0
df = pd.concat(dfs, axis=1).fillna(0)
for col in df.columns:  # Skip first column
    df[col] = df[col].astype(int)
df = df.sort_index(axis=1).reset_index()
df.rename(columns={gene_column: 'Gene'}, inplace=True)
# df
df.to_csv(folder + '/counts.txt', index=False)
```

```python
# THIS IS FOR EXPERIMENTS WITH MULTIPLE FILES, WHERE EACH FILE HAS MULTIPLE SAMPLES
# As if every file was in the same format as the one we need, but split in multiple files for some reason

import glob

# Get a list of CSV files (modify the pattern if needed)
files = glob.glob(folder + "/*.txt")  # Change path accordingly

# Initialize merged DataFrame
df = None

for file in files:
    print(file)
    # filename = os.path.splitext(os.path.basename(file))[0].split('_')[1].replace('Counts', '')
    local_df = pd.read_csv(file, sep='\t', dtype={"Gene": str})
    # print(local_df)
    local_df = local_df.rename(columns={col: col.split('/')[-1].split('.')[0][:-7] for col in local_df.columns if col != "Gene"})
    if df is None:
        df = local_df  # First file, set as base
    else:
        df = df.merge(local_df, on="Gene", how="outer")  # Merge on "Gene" with outer join

df = df.fillna(0).set_index('Gene')
df = df.sort_index(axis=1).reset_index()
df.to_csv(folder + '/counts.txt', index=False)
```

```python
# FOR IDENTIFYING GROUPS OF SAMPLES BASED ON THE .soft FILE

df = pd.read_csv(folder + '/counts.txt', sep=',')
ids = []
strains = []
time = []
with open(folder + '/family.soft') as f:
    for line in f:
        if line.startswith('!Sample_characteristics_ch1'):
            data = line.split('\t')[1:]
            if data[0].startswith('"infection:'):
                for col in data:
                    strains.append(col.split()[-1].replace('/', '_').replace('"', ''))
            if data[0].startswith('"time post infection:'):
                for col in data:
                    time.append(col.split()[-1].replace('"', ''))

        if line.startswith('!Series_sample_id'):
            data = line.split('\t')[1].split()
            for col in data:
                ids.append(col.replace('"', ''))

ids = ids[:-1]

with open(folder + '/input.txt', 'w') as f:
    f.write('SampleName\tReplication\tIdentifier\tFile\n')
    for i in range(len(ids)):
        # group = '_'.join(col.split('_')[1:])
        sample = ids[i]
        group = f'{strains[i]}_{time[i]}'
        print(sample)
        print(group)
        f.write(sample + '\t' + sample + '\t' + group + '\t' + sample + '.tar.gz\n')
```
